## Teoría

### Aproximación de funciones por funciones ortogonales complejas

Dada una función f (t), ésta puede aproximarse mediante funciones complejas que
sean ortogonales de manera que
$$
\begin{equation}
f_a(t)=\tilde{\mathscr{D}}_0\mathscr{G}_0(t)+[\tilde{\mathscr{D}}_0\mathscr{G}_0(t)]^*+\tilde{\mathscr{D}}_1\mathscr{G}_1+[\tilde{\mathscr{D}}_1 \mathscr{G}_1]^*+\cdots+\tilde{\mathscr{D}}_n\mathscr{G}_n(t)+[\tilde{\mathscr{D}}_n\mathscr{G}_n(t)]^*,
\end{equation} 
$$

donde $\tilde{\mathscr{D}}_k$ representa a un coeficiente complejo, $\mathscr{G}_k$ a una función compleja y $^*$ simboliza el conjugado de la función compleja. Como nuestra función original es real, se suma el conjugado de la función compleja para que la parte imaginaria sea cero, aunque se obtenga dos veces la parte real.

### Serie Finita de Fourier en su forma compleja

Un conjunto de funciones complejas que cumplen con la condición de ortogonali-
dad son las funciones exponenciales complejas definidas como
$$
\begin{equation}
\mathscr{G}_k(t)=\exp^{\frac{i2\pi kt}{T}}, \tag{2}
\end{equation}
$$
donde $T$ se es el periodo. El conjugado de la ec. (2) está dado por
$$
\begin{equation}
\mathscr{G}_k^*(t)=\exp^{-\frac{i2\pi kt}{T}}. \tag{3}
\end{equation}
$$
La condición de ortogonalidad es
$$
\begin{equation}
\int_{t_1}^{t_1+T}\exp^{\frac{i2\pi kt}{T}}\exp^{-\frac{i2\pi jt}{T}}dt=\left\{\begin{array}{cc}
0 & \textrm{si }k\neq j \\
T & \textrm{si }k=j. \tag{4}
\end{array}\right.
\end{equation}
$$
*Demostración*. La demostración se hará para cada uno de los posibles casos
* Caso $k \neq j$
$$
\begin{align}
\int_{t_1}^{t_1+T}\exp^{\frac{i2\pi kt}{T}}\exp^{-\frac{i2\pi jt}{T}}dt&=\int_{t_1}^{t_1+T}\exp^{\frac{i2\pi(k-j)t}{T}}dt \nonumber \\
& \textrm{haciendo $l=k-j$ y utilizando coordenadas polares} \nonumber \\
&= \int_{t_1}^{t_1+T}\cos\left(\frac{2\pi lt}{T}\right)+i\sin\left(\frac{2\pi lt}{T}\right)dt. \tag {5}
\end{align}
$$
Trabajando con el primer término de la integral (5)
$$
\begin{align}
\int_{t_1}^{t_1+T}\cos\left(\frac{2\pi lt}{T}\right)dt&=\left(\frac{T}{2\pi l}\right)\sin\left.\left(\frac{2\pi lt}{T}\right)\right|_{t_1}^{t_1+T} \nonumber \\
&=\left(\frac{T}{2\pi l}\right)\left(\sin\left(\frac{2\pi l t_1}{T}+2\pi l\right)-\sin\left(\frac{2\pi lt_1}{T}\right)\right) \nonumber \\
&\textrm{Usando la identidad trigonom\'etrica de }\sin(\theta_1+\theta_2) \nonumber \\
&=\left(\frac{T}{2\pi l}\right)\left(\sin\left(\frac{2\pi l t_1}{T}\right)\cos(2\pi l)+\cos\left(\frac{2\pi lt_1}{T}\right)\sin(2\pi l)-\sin\left(\frac{2\pi lt_1}{T}\right)\right) \nonumber \\
&=\left(\frac{T}{2\pi l}\right)\left(\sin\left(\frac{2\pi lt_1}{T}\right)\left[-1+\cos(2\pi l)\right]+\cos\left(\frac{2\pi lt_1}{T}\right)\sin(2\pi l)\right) \nonumber \\
&\textrm{Como }\cos(2\pi l)=1 \textrm{ y }\sin(2\pi l)=0 \nonumber \\
&=0. \nonumber
\end{align}
$$
Trabajando con el segundo término de la integral (5)
$$
\begin{align}
\int_{t_1}^{t_1+T}\sin\left(\frac{2\pi l}{T}t\right)dt&=\left(\frac{T}{2\pi l}\right)\left(-\left.\cos\left(\frac{2\pi l}{T}t\right)\right|_{t_1}^{t_1+T}\right) \nonumber \\
&=\left(\frac{T}{2\pi l}\right)\left(-\cos\left(\frac{2\pi lt_1}{T}+2\pi l\right)+\cos\left(\frac{2\pi lt_1}{T}\right)\right) \nonumber \\
&\textrm{Usando la identidad trigonom\'etrica de }\cos(\theta_1+\theta_2) \nonumber \\
&=\left(\frac{T}{2\pi l}\right)\left(-\left[\cos\left(\frac{2\pi lt_1}{T}\right)\cos(2\pi l)-\sin\left(\frac{2\pi lt_1}{T}\right)\sin(2\pi l)\right]+\cos\left(\frac{2\pi lt_1}{T}\right)\right) \nonumber \\
&=\left(\frac{T}{2\pi l}\right)\left(\cos\left(\frac{2\pi l t_1}{T}\right)[1-\cos(2\pi l)]-\sin\left(\frac{2\pi lt_1}{T}\right)\sin(2\pi l)\right) \nonumber \\
&=0. \nonumber
\end{align}
$$
* Caso $k \neq j$
$$
\begin{align}
\int_{t_1}^{t_1+T}\exp^{\frac{i2\pi jt}{T}}\exp^{\frac{-i2\pi jt}{T}}dt&=&\int_{t_1}^{t_1+T}dt \nonumber \\
&=&T. \nonumber 
\end{align}
$$
Sustituyendo las funciones exponenciales complejas en $fa(t)$ (1) se tiene
$$
\begin{align}
f_a(t)&=\tilde{\mathscr{D}}_0\exp^{\frac{i2\pi 0t}{T}}+\tilde{\mathscr{D}}_0^*\exp^{-\frac{i2\pi 0t}{T}}+\tilde{\mathscr{D}}_1\exp^{\frac{i2\pi 1t}{T}}+\tilde{\mathscr{D}}_1^* \exp^{-\frac{i2\pi 1t}{T}}+\cdots+ \nonumber \\
&\tilde{\mathscr{D}}_n\exp^{\frac{i2\pi nt}{T}}+\tilde{\mathscr{D}}_n^*\exp^{-\frac{i2\pi nt}{T}} \nonumber \\
&=[\tilde{\mathscr{D}}_0+\tilde{\mathscr{D}}_0^*]+\tilde{\mathscr{D}}_1\exp^{\frac{i2\pi t}{T}}+\tilde{\mathscr{D}}_1^* \exp^{-\frac{i2\pi t}{T}}+\cdots+ \nonumber \\
&\tilde{\mathscr{D}}_n\exp^{\frac{i2\pi nt}{T}}+\tilde{\mathscr{D}}_n^*\exp^{-\frac{i2\pi nt}{T}} \nonumber \\
&=2\text{Re}[\tilde{\mathscr{D}}_0]+\tilde{\mathscr{D}}_1\exp^{\frac{i2\pi t}{T}}+\tilde{\mathscr{D}}_1^* \exp^{-\frac{i2\pi t}{T}}+\cdots+ \nonumber \\
&\tilde{\mathscr{D}}_n\exp^{\frac{i2\pi nt}{T}}+\tilde{\mathscr{D}}_n^*\exp^{-\frac{i2\pi nt}{T}}. \tag {6}
\end{align}
$$
Si de manera conveniente hacemos que $\mathscr{D}_{k}=\tilde{\mathscr{D}}_k$, $\mathscr{D}_{-k}=\tilde{\mathscr{D}}_k^*$ y $\mathscr{D}_0=2\text{Re}[\tilde{\mathscr{D}}_0]$, podemos cambiar los límites de la suma (6) de $-N$ a $N$ como
$$
\begin{equation}
f_a(t)=\sum_{k=-N}^N\mathscr{D}_k\exp^{\frac{i2\pi kt}{T}}, \tag{7}
\end{equation}
$$
la cual se conoce como **serie de Fourier**.
Para calcular sus coeficientes, $\mathscr{D}_k$, seguimos el método del error cuadrático medio mínimo (ECMM).
El error cuadrático medio está dado por
$$
\begin{align}
E^2(t)&=\frac{1}{T}\int_{t_1}^{t_1+T}(f(t)-f_a(t))^2 dt \nonumber \\
&=\frac{1}{T}\int_{t_1}^{t_1+T}\left(f(t)-\sum_{k=-N}^N\mathscr{D}_k\exp^{\frac{i2\pi kt}{T}}\right)^2 dt. \tag{8}
\end{align}
$$
Los coeficientes que minimizan al error cuadrático (8) cumple con
$$
\begin{equation}
\frac{\partial E^2}{\partial\mathscr{D}_k}=0\qquad\forall k. \tag{9}
\end{equation}
$$
Considerando la derivada con respecto al $j$-ésimo coeficiente
$$
\begin{align}
\frac{\partial E^2(t)}{\partial\mathscr{D}_j}&=\frac{\partial}{\partial\mathscr{D}_j}\left[\frac{1}{T}\int_{t_1}^{t_1+T}\left(f(t)-\sum_{k=-N}^N\mathscr{D}_k\exp^{\frac{i2\pi kt}{T}}\right)^2 dt\right] \nonumber \\
&=\frac{1}{T}\int_{t_1}^{t_1+T}2\left(f(t)-\sum_{k=-N}^N\mathscr{D}_k\exp^{\frac{i2\pi kt}{T}}\right)\left( -\exp^{\frac{i2\pi jt}{T}}\right)dt \nonumber \\
&\textrm{Utilizando la condici\'on de ortogonalidad , s\'olo sobrevive el t\'ermino }k=-j \nonumber \\
&=\frac{2}{T}\left(\int_{t_1}^{t_1+T}\left(-f(t)\exp^{\frac{i2\pi jt}{T}}\right)dt+\mathscr{D}_{-j}T\right) \nonumber \\
&\Rightarrow \mathscr{D}_{-j}=\frac{1}{T}\int_{t_1}^{t_1+T}f(t)\exp^{\frac{i2\pi jt}{T}}dt, \tag{10}
\end{align}
$$
que podemos escribir como
$$
\begin{equation}
\mathscr{D}_k=\frac{1}{T}\int_{t_1}^{t_1+T}f(t)\exp^{-\frac{i2\pi kt}{T}}dt, \tag{11}
\end{equation}
$$
usando el cambio de variable $k=-j$.

### Serie de Fourier en su primera forma trigonométrica

La primer forma trigonométrica de la serie de Fourier esta dada de la forma:
$$
fa(t) = \mathscr{C}_0 + \sum_{k=1}^N \mathscr{C}_kcos\left(\frac{2\pi kt}{T}+\phi_k\right)
$$
donde<br><br>
$$\mathscr{C}_0 = \mathscr{D}_0$$
$$\mathscr{C}_k = 2\left|\mathscr{D}_k\right|$$
$$\phi_k = tan^{-1}\left(\frac{Im(\mathscr{D}_k)}{Re(\mathscr{D}_k)}\right)


## Ejemplo 1.5 de las notas de Anguiano

Dada $u(t) = t$ en $t\:\epsilon\:(-\pi,\pi)$, obtener su función aproximada mediante la serie de Fourier

### Solución
Se importan las bibliotecas de Python
* *numpy* : Permite crear y gestionar arreglos multidimensionales, junto con una gran colección de funciones matemáticas de alto nivel que operan sobre estos arreglos. El sitio oficial es https://numpy.org/
* *matplotlib* : Permite crear visualizaciones estáticas, animadas e interactivas en 2D y 3D. El sitio oficial es https://matplotlib.org/
* *scipy* : Es una colección de herramientas y algoritmos matemáticos para Python. Esta librería proporciona funciones para realizar integración numérica, es decir, calcular el valor numérico de una integral definida. La función *quad* se utiliza específicamente para calcular la integral numérica de una función en un intervalo definido. El sitio oficial es https://scipy.org/


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

De la ecuación (7) y (11), tenemos que:
$$
\begin{equation}
f_a(t)=\sum_{k=-N}^N\mathscr{D}_k\exp^{\frac{i2\pi kt}{T}}
\end{equation}
$$
$$
\begin{equation}
\mathscr{D}_k=\frac{1}{T}\int_{t_1}^{t_1+T}f(t)\exp^{-\frac{i2\pi kt}{T}}dt
\end{equation}
$$
Sustituyendo la función dada
$$
\begin{equation}
\mathscr{D}_k=\frac{1}{2\pi}\int_{t_1}^{t_1+T}t\:\exp^{-\frac{i2\pi kt}{T}}dt
\end{equation}
$$

```python
#Creamos una constante i puramente imaginaria
i = complex(0,1)
#otra forma de scribir un número imaginario en Python seria:
#i = 0 + 1j
#Donde j es la forma de representar a i
#Creamos el periodo
ti = -np.pi
tf = np.pi
T = tf - ti
#definimos nuestra función original f(t)
def u(t):
    return t
t = np.linspace(ti,tf,100) #"Intervalo de la variable independiente"
```

```python
#Definimos la exponencial compleja del integrando
def exp(t,k):
    return np.e**(-i*k*t)
#Definimos el integrando completo
def integrando(t,k):
    return u(t)*exp(t,k)   
```

```python
n = int(input('Aproximación de puntos Dk:'))
```

```python
# Creamos nuestro arreglo de coeficientes y función aproximada
D = np.zeros([n,1],dtype=complex)
#Arreglos creado para graficar los coeficientes de la primer forma trigonométrica
C = np.zeros([n,1])
phi = np.zeros([n,1])
fa = 0 #Inicializar la función aproximada
# Ciclo para el cálculo de los coeficientes Dk
for k in range(n):
    #quad no funciona con numeros complejos, por lo que se hace por separado
    #Primero la parte real de la integral
    integral_real, _ = quad(lambda t: integrando(t, k).real, ti, tf)
    #Segundo la parte imaginaria de la integral
    integral_imag, _ = quad(lambda t: integrando(t, k).imag, ti, tf)
    #El resultado de la integral sera la suma de la parte real y la imaginaria
    integral = integral_real + 1j*integral_imag
    coef = complex(1/(2*np.pi), 0)  # Definimos el coeficiente como un número complejo
    # Coeficientes de la forma compleja de la Serie de Fourier
    D[k] = integral * coef
    # Coeficientes y fases de la primer forma trigonométrica
    C[k] = 2*abs(D[k])
    phi[k] = np.arctan2(D[k].imag,D[k].real)
    # Aproximación de la función con la primer forma trigonométrica
    fa += C[k]*np.cos((2*np.pi*k*t)/T+phi[k])
#Imprimimos los coeficientes Dk
D
```

Grafica de la función original y la función aproximada por la primer forma trigonométrica

```python
# Se crea la figura
fig = plt.figure()

# Se grafica el vector de la función original con respecto al vector de la variable independiente
plt.plot(t, u(t), label='$f(t)$')

# Se grafica el vector de la función aproximada con respecto al vector de la variable independiente
plt.plot(t, fa, label='$f_a(t)$')

# Se edita la figura
plt.xlabel('$t$', fontsize=12)  # Se escribe la etiqueta del eje de la variable independiente en tamaño de letra 12
plt.suptitle('Comparación de la función original con la aproximada', weight='bold')  # Se escribe el título de la gráfica
plt.grid()  # Se grafica la cuadrícula
plt.legend()  # Se imprime la leyenda

# Mostrar la gráfica
plt.show()
```

```python
# Se crea la figura
fig = plt.figure()

# Ciclo para el cálculo de los coeficientes Dk
for k in range(n):
    # Aproximación de la función con la primer forma trigonométrica
    fa = C[k]*np.cos((2*np.pi*k*t)/T+phi[k])
    
    # Se el k-ésimo componente de la Serie de Fourier
    plt.plot(t, fa, label=f'Componente {k}')

# Se edita la figura
plt.xlabel('$t$', fontsize=12)  # Se escribe la etiqueta del eje de la variable independiente en tamaño de letra 12
plt.suptitle('Componentes de la Serie de Fourier', weight='bold')  # Se escribe el título de la gráfica
plt.grid()  # Se grafica la cuadrícula
plt.legend()  # Se imprime la leyenda

# Mostrar la gráfica
plt.show()
```
